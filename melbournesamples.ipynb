```python
import shapefile 
from shapely.geometry import Polygon, Point
from numpy import random
import numpy
import json
import math
```

```python
sf = shapefile.Reader("/home/mike/Documents/govhack2020/VMADMIN/layer/postcode_polygon")
```

```python
shapes = sf.shapes() 
pccount = len(shapes)
melbournelong = 145.095927
melbournelat = -37.751098
```

```python
x = sf.shape(2)
```

```python
samplesreqd = 100
# per postcode
samplelist = [] 
howmany = 0
```

```python
for pc in range(pccount) :
    s1 = sf.shape(pc)
    samplecount = 0
    lowerlong = s1.bbox[0]
    upperlong = s1.bbox[2]
    lowerlat = s1.bbox[1]
    upperlat = s1.bbox[3]
    avlat = (lowerlat + upperlat) / 2
    avlong = (lowerlong + upperlong) / 2
    howfar = math.sqrt((melbournelat - avlat)**2 + (melbournelong - avlong)**2)
    print(howfar)
    if (howfar > 0.5) :
        continue
        
    howmany = howmany + 1    
    print(howmany)
    geo = s1.points
    poly = Polygon(geo)
    rec = sf.record(pc)
    postcode = rec.POSTCODE

    while samplecount < samplesreqd:
        longitude = random.uniform(lowerlong, upperlong)
        latitude = random.uniform(lowerlat, upperlat)
        if poly.contains(Point(longitude, latitude)):
            #print(rec.POSTCODE)
            samplecount = samplecount + 1
            samplelist.append({'postcode':postcode,'longitude':longitude, 'latitude':latitude})
        
    
```
